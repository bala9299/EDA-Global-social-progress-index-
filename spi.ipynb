### 6.Draw a plot between wellbeing and basic human needs.

```python
plt.figure(figsize=(18,9))
plt.rcParams.update({'font.size':15})
plt.plot(df['basic_human_needs'],marker='x',c='r',ls='--')
plt.plot(df['wellbeing'],marker='o')
plt.title("percentage between wellbeing and basic human being")
plt.xlabel("spi rank")
plt.ylabel('percentage')
plt.legend(['basic human needs','wellbeing']);
```
